```python
import zlib, os, bz2, pprint

CHUNKSIZE = 1000000
def decompress_data(path):
    data = zlib.decompressobj()  
    my_file = open(path, 'rb')  
    buf = my_file.read(CHUNKSIZE)
    while buf:
        out = data.decompress(buf)
        buf = my_file.read(CHUNKSIZE)
    my_file.close()
    return out

data = decompress_data('level21/package.pack')
reverse = False
logs = [' ']
while True:
    try:
        file = open('level21/package2.pack', 'wb')
        file.write(data)
        file.close()
        decompressed = decompress_data('level21/package2.pack')
        msg = ' zlib'
        logs[len(logs)-1] = logs[len(logs)-1] + ' '
        reverse = False
    except(zlib.error):
        try:
            decompressed = bz2.decompress(data)
            msg = ' bz2'
            logs[len(logs)-1] = logs[len(logs)-1] + '#'
            reverse = False
        except:
            try:
                if reverse == False:
                    decompressed = data[::-1]
                    reverse = True
                    msg = ' reverse'
                    logs.append('')
                else:
                    print('Final message:')
                    print(str(decompressed[:200]))
                    break
            except:
                print(len(data))
                print(data[:200])
                break
    data = decompressed
    # print(str(len(data)) + msg)

pprint.pprint(logs)

# http://www.pythonchallenge.com/pc/hex/copper.html
```
